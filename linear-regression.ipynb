```python
df = pd.read_csv("Advertising.csv")
```

```python
# sales is the target here
# TV, radio, nespaper are features/independent variables
df.drop(columns="Unnamed: 0", inplace=True)
```

## Simple Linear Regression
- only 1 feature in X
- 'y' should be continuos

```python
X = df['TV']
y = df['sales']
X, y
```

```python
# y = a + bX
plt.scatter(X,y, marker='o')
plt.xlabel('TV')
plt.ylabel('sales')
```

```python
X.shape
```

```python
# to split the data there is a function
from sklearn.model_selection import train_test_split
```

```python
X
```

```python
X = X.values.reshape(-1, 1)
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
# output from the model --> parameters a and b
# a = y intercept

lr.intercept_
```

```python
# slope value b
lr.coef_
```

```python
# output from Algorithm

#y = 7.239 + 0.046(TV)
```

```python
# prediction - find output for unkonwn data
lr.predict(X_test)
```

## Evaluation Metrics

Commonly used metrics
1. Mean Absolute Error (MAE) - prediction and target difference and take the mean of error
2. Mean Squared Error (MSE) - most widely used with lr. find squared difference betwen actual vaule y and predicted value
3. Root Mean Squared Error (RMSE) - squared root of MSE
4. R-square(R2)
5. Root Mean Squared Log Error

```python
from IPython.display import Image, display
display(Image(filename="/home/sam/my-study/machinelearning-concepts/Screenshot from 2025-08-21 23-00-38.png"))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
# root mean squared error
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# R2 score is also known as coefficeint of determination, provides the info of the goodbness of a model
# determine how closi is y_pred is with y_true
# value of R2 score will be between 0 to 1 
# value of R2 is 1 when all datapoints falls in the regression line
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(y_test, y_pred)
```

```python
# regularization is to reduce overfit
# ir it works with training data and perform poor with test data
# 3 types of regularization - Ridge Regualrization(L2), Lasso and ElasticNet 
```

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
lasso = Lasso()
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
r2_score(y_test, y_pred)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
r2_score(y_test, y_pred)
```
